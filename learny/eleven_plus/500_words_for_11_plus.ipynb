```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw, ImageFont
import numpy as np

def create_82_mask():
    """Create a mask image in the shape of '82' with thicker strokes in square format."""
    # Create a square image for better proportions
    size = 800
    img = Image.new('RGB', (size, size), 'white')
    draw = ImageDraw.Draw(img)
    
    # Draw "82" with thick strokes centered in square canvas
    # Drawing "8" manually with thick lines
    # Outer circles of 8 (positioned more to the left)
    draw.ellipse([80, 200, 280, 400], fill='black')   # Top circle
    draw.ellipse([80, 350, 280, 550], fill='black')   # Bottom circle
    # Inner holes to make it look like an 8 (smaller holes for thicker lines)
    draw.ellipse([120, 240, 240, 360], fill='white')  # Top hole
    draw.ellipse([120, 390, 240, 510], fill='white')  # Bottom hole
    
    # Drawing "2" manually (positioned more to the right)
    # Top horizontal line
    draw.rectangle([380, 200, 680, 280], fill='black')
    # Right vertical line (top part)
    draw.rectangle([600, 200, 680, 375], fill='black')
    # Middle diagonal/horizontal - thicker stroke
    draw.polygon([(380, 350), (680, 350), (680, 430), (460, 430), (380, 490), (380, 410)], fill='black')
    # Bottom horizontal line
    draw.rectangle([380, 470, 680, 550], fill='black')
    
    # Convert to numpy array and create mask
    mask_array = np.array(img)
    # Make white areas transparent (255) and black areas solid (0)
    mask = np.where(mask_array[:, :, 0] < 128, 0, 255).astype(np.uint8)
    
    return mask

def generate_wordcloud(words=sep_19_words):
    """Generate the word cloud using the provided words."""
    
    # Create the mask
    mask = create_82_mask()
    
    # Convert word list to text
    text = ' '.join(words)
    
    # Create WordCloud with adjusted parameters for square format and thicker fonts
    wordcloud = WordCloud(
        width=800, 
        height=800,  # Square dimensions
        background_color='white',
        mask=mask,
        max_words=80,  # Slightly reduced for better fit
        colormap='viridis',
        relative_scaling=0.4,  # Adjusted for thicker appearance
        min_font_size=12,  # Increased minimum font size for thicker appearance
        max_font_size=120,  # Increased maximum font size
        random_state=42,
        prefer_horizontal=0.8,  # More horizontal text
        font_step=2,  # Larger font steps for bolder appearance
        collocations=False  # Avoid word repetitions
    ).generate(text)
    
    # Create the plot with square aspect ratio
    plt.figure(figsize=(10, 10))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title('Word Cloud in the Shape of "82"', fontsize=18, pad=20)
    plt.tight_layout()
    plt.show()
    
    return wordcloud

# Generate the word cloud
wordcloud = generate_wordcloud()

# convert to image and display and save
# wordcloud = wc.generate(text)
wordcloud.to_file("wordcloud_82.png")   
# wordcloud.to_image()
```
